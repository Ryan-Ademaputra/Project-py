Project kecil-kecilan menghitung data menggunakan mean, median, standart dan variance


```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/Tugas_Statistika/Latihan.csv")
```

#**PROPERTI 1**

```python
konstanta = 10;

# Nilai Awal
tipe = "Tipe2";
mean_awal = df[tipe].mean()
median_awal = df[tipe].median()
variance_awal = df[tipe].var()
std_awal = df[tipe].std()

#Nilai Baru
df[tipe + "_baru"] = df[tipe] + konstanta
mean_baru = df[tipe + "_baru"].mean()
median_baru =df[tipe + "_baru"].median()
variance_baru = df[tipe + "_baru"].var()
std_baru = df[tipe + "_baru"].std()



print(f"=== {tipe} ===")
print(f"Mean Awal       : {mean_awal}")
print(f"Median Awal     : {median_awal}")
print(f"variance Awal   : {variance_awal}")
print(f"std Awal        : {std_awal}")
print(f"mean baru       : {mean_baru}")
print(f"median baru     : {median_baru}")
print(f"Variance baru   : {variance_baru}")
print(f"std baru        : {std_baru}")
print()
```
